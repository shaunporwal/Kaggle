Next is Mutual Information. 

Mutual Information between two variables measures the dependence of one variable to another. If X and Y are two variables, and

> a)If X and Y are independent, then no information about Y can be obtained by knowing X or vice versa. Hence their mutual information is 0.

>b)If X is a deterministic function of Y, then we can determine X from Y and Y from X with mutual information 1.

>c)When we have Y = f(X,Z,M,N), 0 < mutual information < 1

**Wrapper Methods**

Wrapper Methods generate models with a subsets of feature and gauge their model performances.
1. Backwards Elimination
2. Recursive Feature Elimination

```python
# b) Wrapper Methods
# (i) Backward Elimination (computationally expensive but do-able)
cols = list(db_dummy.columns)
pmax = 1
while (len(cols)>0):
    p = []
    db3 = db_dummy[cols]
    db3 = sm.add_constant(db3)
    model = sm.OLS(y, db3).fit()
    p = pd.Series(model.pvalues.values[1:],index = cols)      
    pmax = max(p)
    feature_with_p_max = p.idxmax()
    if(pmax>0.05):
        cols.remove(feature_with_p_max)
    else:
        break
selected_features_BE = cols
list(selected_features_BE)
```

**Recursive Feature Elimination**

As the name suggests, this method eliminates worst performing features on a particular model one after the other until the best subset of features are known.

```python
# b) Wrapper Methods
# (ii) Recursive Feature Eliination via linear regression
cols = list(X.columns)
model = LinearRegression()

#Initializing RFE model
rfe = RFE(model, 14)   

#Transforming data using RFE
X_rfe = rfe.fit_transform(X, y)  

#Fitting the data to model
model.fit(X_rfe, y)              
temp = pd.Series(rfe.support_,index = cols)

#Let's see the predicted features
selected_features_rfe = temp[temp==True].index
list(selected_features_rfe)
```

**Embedded Methods**

1. LASSO Linear Regression: performed by adding an extra term to the cost function of Linear Regression. This apart from preventing overfitting also reduces the coefficients of less important features to zero

```python
# c) Embedding Method
#Via Lasso
reg = LassoCV()
reg.fit(X, y)
print("Best alpha using built-in LassoCV: %f" % reg.alpha_)
print("Best score using built-in LassoCV: %f" %reg.score(X,y))
coef = pd.Series(reg.coef_, index = X.columns)
```

```python
print("Lasso picked " + str(sum(coef != 0)) + " variables and eliminated the other " +  str(sum(coef == 0)) + " variables")
```

```python
imp_coef = coef.sort_values()
import matplotlib
matplotlib.rcParams['figure.figsize'] = (18, 16.0)
imp_coef.plot(kind = "barh")
plt.title("Feature importance using Lasso Model")
```

```python
#Obtain List of variables with coeff > 0
las_scores = pd.DataFrame({'Attr':coef.index, 'coeff':coef.values}) #create df
best_las_scores = las_scores[las_scores['coeff'] != 0]
best_las_scores.sort_values(by=['coeff'], ascending=True)

#Show list of selecgted features
selected_features_Lasso = list(best_las_scores.Attr.unique())
selected_features_Lasso
```
